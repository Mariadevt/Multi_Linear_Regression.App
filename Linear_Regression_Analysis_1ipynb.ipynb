```python
#Converting the dataset into a Numpy. (iloc -ilocation)

df = df.iloc[:,:].values
df
```

```python
#Encoding

from sklearn.preprocessing import LabelEncoder
```

```python
#Stored under the name Universoty

university = LabelEncoder ()
```

```python
#Apply LabelEncoder to the specific column 'Target'

df[:,34] = university.fit_transform(df[:, 34])
df
```

```python
#use of one hot encoder all the features and target

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#Create a Transformer and name it as ct that will be used to encode categorical features of the dataset aand will pass through alll

ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [34])], remainder = 'passthrough')
```

```python
X = df[:,0:34]
y = df[:,34]
```

```python
y = ct.fit_transform(df)
```

```python
Y = pd.DataFrame(y)
```

```python
Y.head(2)
```

```python
#splting the dataset into training and testing

from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_test = train_test_split(X,y, test_size =0.2, random_state = 0)
```

```python
#fitting the regression model

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Predicying the results

y_pred = regressor.predict(X_test)
y_pred
```

```python
#Calculate the R squared value

from sklearn.metrics import r2_score
r2_score(Y_test, y_pred)
```
